```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df=pd.read_csv('dailyActivity_merged.csv')
```

```python
df
```

```python
df.drop(columns=["TotalDistance","TrackerDistance","LoggedActivitiesDistance","VeryActiveDistance","ModeratelyActiveDistance","LightActiveDistance","SedentaryActiveDistance"],axis=1,inplace=True)
```

```python
df
```

```python
df.rename(columns={'ActivityDate':'date'},inplace=True)
```

```python
df
```

```python
df['TotalMins']=df.VeryActiveMinutes+df.FairlyActiveMinutes+df.LightlyActiveMinutes+df.SedentaryMinutes
```

```python
df
```

```python
df['TotalHours']=round(df.TotalMins/60)
```

```python
df
```

```python
df.info()
```

```python
df.date=pd.to_datetime(df.date)
```

```python
import datetime as dt 
```

```python
df['DayOfWeek']=df["date"].dt.day_name()
```

```python
plt.figure(figsize=(6,4))
plt.hist(df.DayOfWeek,bins=7,color='blue',width=.6)
plt.grid(True)
plt.xlabel('Day Of Week')
plt.ylabel('Frequence per day ')
plt.title('Frequence')
plt.show()
```

```python
numeric_df = df.select_dtypes(include='number')
sns.heatmap(numeric_df.corr())
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df.TotalSteps,df.Calories,c=df.Calories)
medSteps=7405
medCal=2134
plt.axhline(medCal,color='blue',label='median cal')
plt.axvline(medSteps,color='red',label='median steps')
plt.xlabel('Total Steps')
plt.ylabel('calories ')
plt.legend()
plt.title('Calories per steps')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df.TotalHours,df.Calories,c=df.Calories)
medHours=24
medCal=2134
plt.axhline(medCal,color='blue',label='median cal')
plt.axvline(medHours,color='red',label='median Hours')
plt.xlabel('Total Hours')
plt.ylabel('calories ')
plt.legend()
plt.title('Calories per Hour')
plt.show()
```
